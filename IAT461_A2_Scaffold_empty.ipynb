### A1 — Exploratory Data Analysis

#### A1.1 — Outcome distribution


```python
outcome = df_ab['avg_whiteboard_session_duration_min']

log_outcome = np.log(outcome)

fig, axes = plt.subplots(1,3, figsize = (16,4))

# raw session duration histogram
sns.histplot(outcome, bins=30, kde=True, ax=axes[0], color=ACCENT)
axes[0].set_title('Raw session duration')
axes[0].set_xlabel('Minutes')

# log histogram
sns.histplot(log_outcome, bins=30, kde=True, ax=axes[1], color=ACCENT)
axes[1].set_title("Log session Duration")
axes[1].set_xlabel('Minutes')

# Q-Q plot
stats.probplot(log_outcome, dist="norm", plot=axes[2])
axes[2].set_title("Q-Q plot of log session duration")

plt.tight_layout()
plt.show()

# Compute and output skewness for both raw and log-transformed distributions
raw_skew = outcome.skew()
log_skew = log_outcome.skew()

print(f'raw skew: {raw_skew}')
print(f"log skew: {log_skew}")
```

**✏️ Your interpretation (A1.1):**

The log-transformed outcome is more normally distributed. 

The raw session duration was strongly positively skewed (1.377), indicating that a small amount of users had very long whiteboard sessions. The log transformed session duration had a skewedness that decreased to -0.213, resulting in a distributionthat was much closer to being symmetric. The Q-Q plot shows most points following the reference line, with minor deviations at the start and end.

### A2 — Feature Engineering

Prepare your feature matrix for OLS. Exclude the two problematic variables identified in A1.


```python
BEGIN
    Load baseline dataset

    // 1. Encode Categorical Variables
    Convert experimental group ('ab_group') into a binary integer indicator
    Generate dummy indicators (one-hot encoding) for device types, omitting the first category
    Generate dummy indicators (one-hot encoding) for regions, omitting the first category

    // 2. Perform Optional Transformations
    Log-transform right-skewed numerical predictors using log(x + 1) to avoid zero values

    // 3. Assemble Modeling Dataset
    Merge numerical characteristics, binary groups, and dummy indicators into matrix X
    Set target vector y as the log-transformed outcome variable

    // 4. Quality Control
    Verify that there are no missing (null) values in feature matrix X
    Inspect matrix dimensions and print a preview of the processed data
END
```

```python
# load baseline
df_model = df_ab.copy()

# ecnode categorical
df_model['ab_group'] = (
    df_model['ab_group'] == 'treatment'
).astype(int)

# one hot encode categoricals
# omit first
device_dummies = pd.get_dummies(
    df_model["device_type"], prefix="device", drop_first=True, dtype=int
)

region_dummies = pd.get_dummies(
    df_model["plan_region"], prefix="region", drop_first=True, dtype=int
)

skewed_predictors = [
    'prior_whiteboard_sessions',
    'num_collaborators',
    'templates_used'
]

# log each with log1p
for col in skewed_predictors:
    df_model[f"log_{col}"] = np.log1p(df_model[col])

# matrix x

X = pd.concat(
    [
        df_model[
            [
                "ab_group",
                "account_age_days",
                "feature_tour_completed",
                "log_prior_whiteboard_sessions",
                "log_num_collaborators",
                "log_templates_used",
            ]
        ],
        device_dummies,
        region_dummies
    ],
    axis=1
)

df_model["log_session_duration"] = np.log(
    df_model["avg_whiteboard_session_duration_min"]
)

y = df_model["log_session_duration"]

print(X.isnull().sum())
print(X.shape)
X.head()
```
